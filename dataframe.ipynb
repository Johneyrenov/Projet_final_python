### Reinitialiser l'index

```python
df.reset_index(drop=True,inplace=True)
df
```

```python
age_order = ['Moins de 18 ans', '18 - 25 ans', '25 - 30 ans', '30 - 40 ans', '40 et plus']
df['age'] = pd.Categorical(df['age'], categories=age_order, ordered=True)
```

```python
df['sexe'] = df['sexe'].astype('category')
```

```python
df['plateforme_preferee'] = df['plateforme_preferee'].astype('category')
```

```python
df['appareils_utilises'] = df['appareils_utilises'].astype('string')
```

```python
df['temps_reseaux'] = df['temps_reseaux'].astype('string')
```

```python
time_mapping = {
    "moins de 15 minutes": 7.5, # (0 + 15)/2
    "15 minutes - 45 minutes": 30, #(15+45)/2
    "45 minutes - 2 heures": 82.5, #(45+120)/2
    "2 heures - 4 heures": 180, #(120 + 240)/2
    "4 heures - 8 heures": 360, #(240 + 480)/2
    "plus de 8 heures": 600 #10h ou 12h si necessaire
}

df["temps_reseaux_mns"] = df["temps_reseaux"].map(time_mapping)
df
```

```python
df['temps_streaming'] = df['temps_streaming'].astype('string')
```

```python
df["temps_streaming_mns"] = df["temps_streaming"].map(time_mapping)
df
```

```python
df['temps_jeux'] = df['temps_jeux'].astype('string')
```

```python
df["temps_jeux_mns"] = df["temps_jeux"].map(time_mapping)
df
```
